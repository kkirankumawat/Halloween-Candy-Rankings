## Using Clustering on the relevant variables 

```python
df.head()
```

```python
df_new = df[['competitorname','chocolate', 'fruity', 'peanutyalmondy', 'winpercent']]
df_new.head()
```

```python
df1_new = df_new.iloc[:,1:-1]
df1_new.head()
```

```python
import sklearn
from sklearn.preprocessing import StandardScaler 

scaler = StandardScaler()
df1_new_scaled = scaler.fit_transform(df1_new)
df1_new_scaled
```

```python
df1_new_scaled.shape
```

```python
df1_new_scaled = pd.DataFrame(df1_new_scaled)
df1_new_scaled.head()
```

```python
df1_new_scaled.columns = ['chocolate', 'fruity', 'peanutyalmondy']
df1_new_scaled.head()
```

### Finding optimum number of clusters

```python
from sklearn.cluster import KMeans
```

```python
# elbow curve method (sum of squared distances)
ssd=[]
range_n_clusters=[2,3,4,5,6,7,8]
for num_clusters in range_n_clusters:
    kmeans=KMeans(n_clusters=num_clusters, max_iter=100)
    kmeans.fit(df1_new_scaled)
    ssd.append(kmeans.inertia_)
    
# plot the SSDs for each n_clusters
plt.plot(ssd)
plt.show()
```

```python
# silhouette analysis
from sklearn.metrics import silhouette_score

range_n_clusters=[2,3,4,5,6,7,8]
for num_clusters in range_n_clusters:
   
    # initialise k means
    kmeans=KMeans(n_clusters=num_clusters, max_iter=100)
    kmeans.fit(df1_new_scaled)
    
    cluster_labels=kmeans.labels_
    
    # silhouette score
    silhouette_avg=silhouette_score(df1_new_scaled, cluster_labels)
    print('For n_clusters={0}, the silhouette score is {1}'.format(num_clusters, silhouette_avg))
```

```python
kmeans=KMeans(n_clusters=3, max_iter=50)
kmeans.fit(df1_new_scaled)
```

```python
kmeans.labels_
```
